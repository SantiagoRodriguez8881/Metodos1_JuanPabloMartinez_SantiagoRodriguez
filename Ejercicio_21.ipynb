```python
# EJERCICIO 21

import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import math


# b)) Implementar la funncion sgn(x)
@np.vectorize()
def sgn(x):
  if x > 0:
    return 1
  elif x == 0:
    return 0
  else:
    return -1

# c)) Vectorizar y pintar la funcion sgn(x)

X = np.linspace(-1,1,90)
Y = np.vectorize(sgn)(X)
plt.scatter(X,Y,color = 'r')
plt.scatter(0,0,c = 'r')
plt.title("Funcion signo")
plt.grid()
plt.show()

# d)) Definir los puntos y los pesos de Gauss-Legendre al grado n= 15

puntos, pesos = np.polynomial.legendre.leggauss(15)
print(puntos,pesos)

# e)) Crear y guardar los polinomios de Legendre hasta N = 20

x = sym.Symbol('x',real = True)
def Getpols(n,x):
  if n == 0:
    return 1
  elif n == 1:
    return x
  else:
    P = ((x**2)-1)**n
    for i in range(n):
      dP = sym.diff(P,x)
      P = dP
    return (1/((2**n)*math.factorial(n))) * dP

Pols = []
Pols.append(Getpols(0,x))
for i in range(1,21):
    Pols.append(Getpols(i,x))

print(Pols)

# f)) Creemos una funcion que calcule los N+ 1 coeficientes en un array de numpy:

def GetCoeff(n,f,Pols):
  Cn = 1.0
  if Pols[n] == 1:
    Cn *= np.sum(pesos*f(puntos))
  else:
    Cn *= np.sum(pesos*f(puntos)* np.array([Pols[n].subs(x, p) for p in puntos], dtype=float))
  Cn *= (2*n + 1) /2.
  return Cn
GetCoeff(5,sgn,Pols)
Cn = []
for i in range(21):
  Cn.append(GetCoeff(i,sgn,Pols))
print(Cn)


# g)) Vamos a calcular la combinacion lineal segun la expresion 4.195 :::

def Combination(Pols, n, Cn1, X):
  Y1 = np.zeros_like(X)
  for i in range(len(X)):
    suma = 0
    for j in range(n+1):
      if Pols[j] == 1:
        suma += Cn1[j]
      else:
        suma += Cn1[j]*Pols[j].subs(x,X[i])
    Y1[i] = suma
  return Y1

# Vamos a pintar la funcion signo y la aproximacion con el polinomio de Legendre

Y1 = Combination(Pols,20,Cn,X)
plt.plot(X,Y1)
plt.scatter(X,Y,c = 'r')
plt.scatter(0,0,color = 'r')
plt.title("Expansion de la base de legendre al grado N = 20")
plt.xlabel("")
plt.grid()
plt.show()
```
