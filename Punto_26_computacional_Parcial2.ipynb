```python
# Ejercicio 26, parte computacional
import numpy as np
import matplotlib.pyplot as plt
@np.vectorize()

# e) Definamos una funcion que aproxime la integral con n=30 subintervalos:

def Suma_Riemman(a,b,f,n):
  delta = (b-a)/n
  suma = 0
  for i in range(n):
    suma += f(a+i*delta)*delta
  return suma

def f(x):
  return x**3

print(Suma_Riemman(0,2,f,30)) # Nos podemos dar cuenta que con 30 puntos no basta para hacer una gran aproximacion, pero no está mal

# f) Vamos a dibujar el error de la aproximacion en funcion de n:

# Para esto quiero definir la funcion de la integral exacta:

def f_exacta(x):
  return x**4/4

def error(n):
  return np.abs(f_exacta(2) - Suma_Riemman(0,2,f,n))

n_nuevo = np.linspace(30,400,100, endpoint = False, dtype= int) # Ajuste el linspace para que sea aceptado a la hora de evaluar el array en la suma de riemman
versus = error(n_nuevo)

plt.plot(n_nuevo, versus)
plt.xlabel("n")
plt.ylabel("Error")
plt.show()

# g)

# Justificacion: La gráfica nos está mostrando el error de la estimacion, esto nos puede ayudar a analizar que tan bueno puede ser
# usar este metodo para aproximar integrales. Si estamos buscando un metodo sencillo para aproximarlas siento que es adecuado, una
# desventaja que le encuentro es que requiere de muchos puntos para tener una presicion de menos del orden de 1e-2, lo cual para problemas
# mas complicados puede resultar en algo desventajoso, ya que puede que la suma sea muy costosa o se requiera mucha precision. Frente a lo
# aprendido sobre la cuadratura Gaussiana, siento que la cuadratura es un mejor metodo ya que ofrece una convergencia mucho mayor y datos
# mas precisos con una cantidad relativamente pequeña de puntos y pesos, por ejemplo con 10 pesos la precision es inmensa, es por esto que
# opino que la cuadratura Gaussiana es un mejor metodo.
```
