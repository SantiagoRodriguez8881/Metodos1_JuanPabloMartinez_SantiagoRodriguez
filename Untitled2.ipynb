```python
import numpy as np
import sympy as sym
import matplotlib.pyplot as plt

#Vamos a crear el conjunto de soporte

X = np.array([1.4, 3.5, 5.6])
Y = np.array([0.4007954931819738, 0.594128102489774, 0.29802795523938164])
x = sym.Symbol('x',real=True)

#Creamos el algoritmo que sintetiza la productoria dentro de la interpolacion
def Lagrange(x,X,i):
  c = 1
  for j in range(len(X)):
    if i != j:
      c *= (x-X[j])/(X[i]-X[j])
  return c

#Creamos el algoritmo que sintetiza la interpolacion de Lagrange
def Aproximacion(x,X,Y):
  suma = 0
  for i in range(len(X)):
    suma += Y[i]*Lagrange(x,X,i)
  return suma

#Vamos a ver que funcion produce la interpolacion a los puntos en X y en Y
b = Aproximacion(x,X,Y)
b = b.expand()
b
```

```python
import numpy as np
from scipy.interpolate import lagrange
import matplotlib.pyplot as plt

# Datos proporcionados
x_data = np.array([1.4, 3.5, 5.6])
y_data = np.array([0.4007954931819738, 0.594128102489774, 0.29802795523938164])

# Interpolación de Lagrange
polynomial = lagrange(x_data, y_data)

# Crear una función para la parábola ajustada
def trajectory(x):
    return polynomial(x)

# Calcular los coeficientes del polinomio de segundo grado
# La forma del polinomio de Lagrange es general, necesitamos ajustar sus coeficientes
coefficients = polynomial.coef
a = coefficients[2]  # Coeficiente de x^2
b = coefficients[1]  # Coeficiente de x
c = coefficients[0]  # Término constante

# Ajustar a la forma y = a*x - b*x^2
a_adjusted = -a
b_adjusted = -b

# Constantes
g = 9.8  # Aceleración debida a la gravedad en m/s^2

# Calcular el ángulo theta
theta = np.arctan(a_adjusted)  # Ángulo en radianes
theta_deg = np.degrees(theta)  # Convertir a grados

# Calcular la velocidad inicial V0
cos_theta_squared = 1 / (1 + a_adjusted**2)
V0_squared = -g / (2 * b_adjusted * cos_theta_squared)
V0 = np.sqrt(V0_squared)

# Mostrar resultados
print(f"Ángulo de lanzamiento: {theta_deg:.2f} grados")
print(f"Velocidad inicial: {V0:.2f} m/s")

# Graficar los datos y la parábola ajustada
x_fit = np.linspace(min(x_data), max(x_data), 100)
y_fit = trajectory(x_fit)

plt.figure(figsize=(8, 6))
plt.scatter(x_data, y_data, color='red', label='Datos experimentales')
plt.plot(x_fit, y_fit, label='Interpolación de Lagrange', color='blue')
plt.xlabel('x (m)')
plt.ylabel('y (m)')
plt.title('Interpolación de Lagrange para la trayectoria de un proyectil')
plt.legend()
plt.grid(True)
plt.show()
```
